Notes: Different order of color channels

```
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(imgrgb)
plt.show()

# Or, use fun numpy functions / indexing!
imgrgb = img[:,:,::-1]
plt.imshow(imgrgb)
plt.show()
```

##Accessing pixels (Numpy Basic)

```
# Get value of a specific pixel
px = imgrgb[100,100]

print(px)

# Change pixel value
imgrgb[55,55] = [255,255,255]
```

```
# Show specific region of the image
plt.axis("off")
plt.imshow(imgrgb[0:200,250:500,:])
plt.show()
```

##Basic Operations on Images

```
# Resize to specific size
res = cv2.resize(imgrgb, (150,100))
plt.imshow(res)
plt.show()

# Expanding by scaling
res = cv2.resize(imgrgb, None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
plt.show()

# Shrinking by scaling
res = cv2.resize(imgrgb, None,fx=.1, fy=.1, interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
plt.show()
```

```
img_flip_x = cv2.flip(imgrgb, 0)  # flip along its horizontal x-axis
plt.axis("off")
plt.imshow(img_flip_x)
plt.show()

img_flip_y = cv2.flip(imgrgb, 1)  # flip along its vertical y-axis
plt.axis("off")
plt.imshow(img_flip_y)
plt.show()

img_flip = cv2.flip(imgrgb, -1)   # how about this one?
plt.axis("off")
plt.imshow(img_flip)
plt.show()
```

##Converting Color Space

```
img = cv2.imread(os.path.join(image_path,'color_wheel.jpg'),cv2.IMREAD_COLOR)
print('Shape:', img.shape)

# Convert BGR to RGB
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(imgrgb)
plt.show()

# Convert BGR to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.axis("off")
plt.imshow(img_gray,'gray')
plt.show()

# Convert BGR to HSV
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.axis("off")
plt.imshow(img_hsv)
plt.show()

# Display H channel
plt.axis("off")
plt.imshow(img_hsv[:,:,0])
plt.show()

# Display S channel
plt.axis("off")
plt.imshow(img_hsv[:,:,1])
plt.show()

# Display V channel
plt.axis("off")
plt.imshow(img_hsv[:,:,2])
plt.show()
```
